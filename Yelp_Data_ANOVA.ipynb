```python
#basic libraries for linear algebra and data procesing
import numpy as np
import pandas as pd

#visualization
import matplotlib.pyplot as plt
import seaborn as sns

#sentiment analysis
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

#comparison analysis metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay

#time and warnings
import time
import warnings

from scipy.stats import f_oneway

#settings
warnings.filterwarnings("ignore")
%matplotlib inline
sns.set_context('poster', font_scale=0.5)
```

```python
yelp_data = pd.read_csv('/Users/machome/OneDrive - UVA/DS-3001/data/Yelpdata.csv')
```

```python
yelp_data.info()
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
yelp_data['polarity_scores'] = yelp_data['text'].apply(lambda text: sid.polarity_scores(str(text)))
yelp_data['compound_score']  = yelp_data['polarity_scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
yelp_data.tail(20)
```

```python
yelp_data.columns
```

```python
# Only need Cat and Composite_Score Columns for ANOVA Test
anovadf = yelp_data.drop(columns=['stars', 'text', 'polarity_scores'], axis=1)
```

```python
anovadf.head(20)
```

```python
# Extract mean values for each category
category_means = [anovadf[anovadf['cat'] == cat]['compound_score'] for cat in anovadf['cat']]

# Perform ANOVA test
f_statistic, p_value = f_oneway(*category_means)

# Output results
print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

```python
# How many reviews do we have per category?
average_count = anovadf.groupby(['cat']).count()
average_count
```

```python
# Average Score Per Category
average_scores = anovadf.groupby(['cat']).mean()
average_scores
```

```python
# Restoring Groupby Output into Dataset
test_df = average_scores.reset_index()
test_df
```
